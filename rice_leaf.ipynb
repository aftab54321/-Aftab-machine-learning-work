```python
import os
os.getcwd()
os.listdir('data_leaf')
os.listdir('data_leaf\Bacterial leaf blight')
```

```python
###tranformation
class Transform:
    def __init__(self,dirs:list[str]) -> None:
        self.from_path='data_leaf'
        self.to_path='processed_data'
        self.dirs=dirs
        self.cls={dirs[i]:i for i in range(len(dirs))}
        self.images_path=[
            (f'data_leaf\\{dir}\\{img}',f'processed_data\\{img.split(".")[0]}_{self.cls[dir]}.{img.split(".")[1]}') for dir in self.dirs for img in os.listdir(f'data_leaf\\{dir}')]
        print(len(self.images_path))  
    
    def transform(self):
        for img,pimg in self.images_path:
            image=io.imread(img,as_gray=True)
            scimg=rescale(image,1/2)
            resize_img=resize(scimg,(228,228))
            io.imsave(pimg,resize_img)
        return True    
```

```python
for dir in ['Bacterial leaf blight','Brown spot','Leaf smut']:
    for img in os.listdir(f'data_leaf\\{dir}'):
        img_path=img.split('.')
        img_path=img_path[0]+f'_0.'+img_path[1]
        print((f'data_leaf\\{dir}\\{img}',f'processed_data\\{img_path}'))
```

```python
tf_image=Transform (['Bacterial leaf blight','Brown spot','Leaf smut'])
```

```python
tf_image.transform()
```

```python
import warnings 
warnings.filterwarnings(action='ignore')
```

```python
files= os.listdir(r'processed_data')
```

```python
images=[]
labels=[]
for _,pimg in tf_image.images_path:
    image=io.imread(pimg).ravel()/255
    label=int(pimg.split('.')[0][-1])
    images.append(image)
    labels.append(label)
```

```python
##logistic regression
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(images,labels,test_size=0.25,random_state=3)
```

```python
xtrain=np.array([image for image in x_train])
ytrain=np.array(y_train)
xtrain[0]
```

```python
from sklearn.linear_model import LogisticRegression

modelL=LogisticRegression(random_state=1)
modelL.fit(xtrain,ytrain)
```

```python
xtest=np.array([image for image in x_test])
ytest=np.array(y_test)
ypred=modelL.predict(x_test)
ypred
```

```python
from sklearn.metrics import plot_confusion_matrix,f1_score,accuracy_score

plot_confusion_matrix(modelL,xtest,ytest)
```

```python
f1_score(y_test,ypred,average='weighted')
```

```python
accuracy_score(ytest,ypred)
```

```python
##svm
```

```python
from sklearn.svm import SVC

svm=SVC(kernel='linear',random_state=42)
svm.fit(xtrain,ytrain)
ypred=svm.predict(xtest)
ypred==ytest
```

```python
plot_confusion_matrix(svm,xtest,ytest)
```

```python
f1_score(y_test,ypred,average='weighted')
```

```python
accuracy_score(ytest,ypred)
```
